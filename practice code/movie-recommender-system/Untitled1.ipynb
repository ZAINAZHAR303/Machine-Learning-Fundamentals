```python
import pandas as pd
from scipy.spatial.distance import cosine

# 1. Create a User-Item Matrix (Ratings 1-5)
data = {
    'Action_Movie': [5, 4, 1, 0],
    'Comedy_Movie': [1, 0, 5, 4],
    'SciFi_Movie':  [5, 5, 2, 1]
}
df = pd.DataFrame(data, index=['User_A', 'User_B', 'User_C', 'User_D'])

# 2. Function to calculate similarity (1 = identical tastes, 0 = no similarity)
def get_similarity(user1, user2):
    # Cosine distance measures the 'angle' between two users' rating vectors
    return 1 - cosine(df.loc[user1], df.loc[user2])

# 3. Compare User A with others
sim_A_B = get_similarity('User_A', 'User_B')
sim_A_C = get_similarity('User_A', 'User_C')

print(f"Similarity between A and B: {sim_A_B:.2f}") # High: Both like Action/SciFi
print(f"Similarity between A and C: {sim_A_C:.2f}") # Low: C prefers Comedy
```
