```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
m_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']

# Load the file with the correct encoding
movies = pd.read_csv('u.item', 
                     sep='|', 
                     names=m_cols, 
                     usecols=range(5), 
                     encoding='latin-1')  # Use 'iso-8859-1' if latin-1 doesn't work

print(movies.head())
print(movies.isnull().sum())
print(movies.shape)
```
